```python


def generate_embedding_matrix(embedding_file, input_file, emb_size = 50):
    vocab = Vocab()
    vecs = [np.random.normal(scale =.5, size = (emb_size))]
    print(len(vecs))
    with open(embedding_file, 'r', encoding='utf-8') as f:
        for line in tqdm(f):
            l = line.split()
            vocab.add(l[0])
            vecs.append(np.array(l[1:]).astype(float))
    
    glove_words = len(vecs)
    print("GLove words:", glove_words)
    vocab = add_file_to_vocab(vocab, input_file)
    new_words = len(vocab) - glove_words
    print("new words:", new_words)
    for i in range(new_words):
        vecs.append(np.random.normal(scale =.5, size = (emb_size)))
    print(len(vecs), len(vocab))
    vocab.add("<UNK>")
    vecs.append(np.random.normal(scale =.5, size = (emb_size)))
    print(len(vecs), len(vocab))
    return vocab, vecs
        
```

```python
v, vecs = generate_embedding_matrix("utils/glove.6B.50d.txt", "Data/train.tsv")
```
